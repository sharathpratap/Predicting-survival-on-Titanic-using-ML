```python
df["Cabin"].replace(np.nan,0,inplace=True)
```

```python
df.drop(["Name","Ticket","Fare","Embarked"],inplace=True,axis = 1)
df
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df.Age.mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
temp = pd.get_dummies(df["Sex"])
temp
```

```python
df = pd.concat([df,temp],axis=1)
df
```

```python
df.drop(["Sex","Cabin"],inplace = True,axis = 1)
df
```

```python
Y = df["Survived"]
features = df.drop(["Survived"],axis=1)
X= features
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=5,random_state=1)
clf.fit(X_train,Y_train)
pred = clf.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(Y_test,pred)
acc
```
